```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

from soccer import Soccer, Player

import CorrelatedQ, FoeQ, FriendQ, Q
```

```python
%matplotlib inline
```

```python
# default value

NUM_ITER = 500000
GAMMA = 0.9

ALPHA_START = 0.1
ALPHA_END = 0.001
ALPHA_DECAY = 0.99999

EPSILON_START = 1.0
EPSILON_END = 0.01
EPSILON_DECAY = (EPSILON_START - EPSILON_END) / NUM_ITER
```

```python
SEED = 11100
```

```python
A, B = Player('A'), Player('B')
```

# Q

```python
game = Soccer(A, B, seed=SEED)
agent = Q.Learner(seed=SEED, alpha_decay=0.999995)
agent.learn(game)
```

```python
plt.plot(agent.q_diff, linewidth=0.5)
plt.ylim(0, 0.5)
plt.xlabel('Simulation Iteration')
plt.ylabel('Q-value Difference')
plt.title('Q Learner')
plt.savefig('q.png')
plt.show()
```

# Friend Q

```python
game = Soccer(A, B, seed=SEED)
agent = FriendQ.Learner(seed=SEED)
agent.learn(game)
```
